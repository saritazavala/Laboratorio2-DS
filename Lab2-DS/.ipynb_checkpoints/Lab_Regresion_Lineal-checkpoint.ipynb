### Revision de VIF (Variance inflation Factor)

```python
to_check = clean(carros, method = 'dropcols', columns = ['price'])
to_check
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["feature"] = to_check.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(to_check.values, i)
                          for i in range(len(to_check.columns))]
vif_data
```

### Re-ordenar un poco

```python
price = carros.price
carros.drop(['price'],axis=1,inplace=True)
carros = pd.concat([price,carros],axis=1)
carros.head()
```

## Modelo de regresión lineal

### Declarar las entradas y las metas

```python
carros.columns
```

```python
X = carros[['brand', 'body','mileage','enginev','engine_type','year','is_registered']]
Y = carros['price']
```

### Escalar (poner a escala) los datos

```python
from sklearn.preprocessing import StandardScaler
escalador = StandardScaler()
```

```python
escalador.fit(X)
cols_std = escalador.transform(X)
```

```python
df_std = pd.DataFrame(cols_std, columns=X.columns)
df_std.head()
```

### Divisón del df completo en datos de entrenamiento y de prueba (Train Test Split)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_std, Y, test_size=0.4, random_state=101)
```

### Crear la regresión

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

### Encontrar R^2 del modelo

```python
# compute with statsmodels, by adding intercept manually
r_cuadrado =  lm.score(df_std,Y)
print("R-cuadrado:", r_cuadrado)
```

Nótese que este no es el R^2 ajustado.  En otras palabras hay que encontrar el R^2 ajustado para tener una medida más adecuada

### Fórmula para R^2 Ajustado

$R^2_{ajust.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

n = número de observaciones

p = número de predictores

```python
r_cuad_aj = 1 - (1-r_cuadrado)*(len(Y)-1)/(len(Y)-df_std.shape[1]-1)
print("R-cuadrado Ajustado", r_cuad_aj)
```

### Encontrar los pesos (coeficientes) y el sesgo (intercepto)

```python
coeficientes = pd.DataFrame(lm.coef_, df_std.columns, columns=['Coeficiente'])
coeficientes
```

## Prueba del modelo

```python
predicciones = lm.predict(X_test)
```

### Completo

```python
plt.xlabel("Real Price")
plt.ylabel("Predicted Price")
plt.scatter(y_test,predicciones)
```

### Para mejor visualizacion

```python
plt.xlim(0, 45500)
plt.ylim(0, 35000)
plt.xlabel("Real Price")
plt.ylabel("Predicted Price")
plt.scatter(y_test,predicciones)
```

```python
sns.displot((y_test - predicciones),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predicciones))
print('MSE:', metrics.mean_squared_error(y_test, predicciones))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicciones)))
```

### Conclusiones


- Todas la variables del dataset fueron utilizadas para crear el modelo exitoso, exceptuando modelo, ya que era un dato cualtitativo con demasiadas categorias por lo que se dificultaba crear variables dummies.
- En nuestro caso obtuvimos un mejor resultado al quitar las filas con datos faltantes en vez de rellenarlos, ya que al rellenarlos podiamos sin querer agregar un dato atipico y complicar nuestro modelo.
- Al quitar los datos atipicos se obtuvo un mucho mejor modelo, incialmente no los habiamos removido y obtuvimos un R-cuadrado de 0.38, bastante bajo para nuestros estandares.
- Tras obtener nuestras variables dummy y realizar un mapa de calor para la correlacion observamos que year e is_registered tienian un poco de correlacion entre ellas, pero no superaba el 0.4 por lo que las decidimos mantener.

- Analizando las metricas podemos observar que obtuvimos Error Absoluto Medio de 4203.59, un Error Cuadrado Medio de 33,034,614.03 y un Raís del Error Cuadrado Medio 5747.57, siendo el mas alto el MSE, este indicador "castiga" los errores mas grandes.
